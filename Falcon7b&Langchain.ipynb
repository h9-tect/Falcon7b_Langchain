# load the tiiuae/falcon-7b-instruct the top model in huggingface leaderboard






```python
from transformers import AutoTokenizer
from torch import cuda, bfloat16

model = "tiiuae/falcon-7b-instruct"

tokenizer = AutoTokenizer.from_pretrained(model)

device = f'cuda:{cuda.current_device()}' if cuda.is_available() else 'cpu'
```

## The provided code sets up a text generation pipeline using the Transformers library. It configures a pre-trained language model, tokenizer, and various parameters for generating text based on the given inputs.

```python
import torch
import transformers

pipeline = transformers.pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
    torch_dtype=torch.bfloat16,
    trust_remote_code=True,
    device_map="auto",
    max_length=200,
    do_sample=True,
    top_k=10,
    num_return_sequences=1,
    eos_token_id=tokenizer.eos_token_id,
    pad_token_id=tokenizer.eos_token_id,
)
```

# imports the HuggingFacePipeline class from the langchain library. This class is used to create a local pipeline from a Hugging Face model. The pipeline variable is a pre-trained model that has been loaded from the Hugging Face Model Hub. The HuggingFacePipeline class takes the pipeline variable as input and creates a local pipeline object that can be used for inference. then assigns the HuggingFacePipeline object to the llm variable. This variable can now be used to interact with the local pipeline. For example, you could use the llm variable to generate text, translate languages, or answer questions.

```python
from langchain import HuggingFacePipeline

llm = HuggingFacePipeline(pipeline=pipeline)
```

```python
from IPython.core.display import display, HTML


def display_result(question, result):
    display(HTML(f"<h3>{question}</h3><p>{result}</p>"))
```

```python
question = "write a story about a prrince"
result = llm(question)
display_result(question, result)
```

# The code imports two classes from the langchain library to create a prompt template and an LLM chain The prompt template has one input variable, instruction, and the template string is "{instruction}".The LLM chain is used to generate text based on instructions, such as generating code, translating languages, or answering questions.


```python
from langchain import PromptTemplate, LLMChain
from langchain.llms import HuggingFacePipeline

# template for an instruction with no input
prompt = PromptTemplate(
    input_variables=["instruction"],
    template="{instruction}"
)

llm = HuggingFacePipeline(pipeline=pipeline)

llm_chain = LLMChain(llm=llm, prompt=prompt)
```

```python
print(llm_chain.predict(
    instruction="Explain to me the difference between nuclear fission and fusion."
).lstrip())
```

```python

from langchain.chains.conversation.memory import ConversationBufferWindowMemory

memory = ConversationBufferWindowMemory(
    memory_key="history",  # important to align with agent prompt (below)
    k=5,
    #return_messages=True  # for conversation agent
    return_only_outputs=True  # for conversation chain
)
```

```python
from langchain.chains import ConversationChain

chat = ConversationChain(
    llm=llm,
    memory=memory,
    verbose=True
)
```

```python
res = chat.predict(input='hi how are you?')
res
```
